```python
from transformers import pipeline
import torch
```

```python
my_gpt_neo_generator =pipeline("text-generation",model="EleutherAI/gpt-neo-125M")
```

```python
promt="How tomake pizza"
res=my_gpt_neo_generator(promt,max_length=200,do_sample=True,temperature=0.9)
print(res[0]["generated_text"])
```
